#Tratamento de Dados com as Notas Objetivas

```python
# Initialize an empty list to store DataFrames
df_list = []

# Iterate through pages 45 to 53
for page in range(45, 54):
    try:
        # Read tables from each page
        tables = tabula.read_pdf("resultadoObjetiva_1.pdf", pages=str(page))

        # Iterate through each table on the page
        for table in tables:
            # Convert the table to a DataFrame
            df = pd.DataFrame(table)
            # Append to the list
            df_list.append(df)
    except Exception as e:
        print(f"Error processing page {page}: {e}")

# Concatenate all DataFrames in the list
if df_list:
    df_tabela_obj = pd.concat(df_list, ignore_index=True)
    print(df_tabela_obj.head()) # Print the first few rows
else:
    print("No tables found in the specified pages.")
```

```python
#col_names = ['Num de inscrição', 'Nome Candidato', 'Vaga', 'Cargo', 'Notas', 'NPD', 'Situação']
col_names = ['Código\rdo Cargo', 'Ocupação', 'Incrição', 'Nome do Candidato','Solicitou\rconcorrer', 'Data de\rNascimento', 'Tipo de\rProva',
       'Pontuação final nas Áreas de Conhecimento', 'Nota Final', 'Situação', 'Classificação',
       'Convocado para correção\rda prova\rdiscrusiva/redação', 'Unnamed: 0', 'Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5',
       'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8']
df_resposta_obj = pd.DataFrame(columns=col_names)
df_resposta_obj = df_tabela_obj.drop(['Solicitou\rconcorrer', 'Data de\rNascimento'], axis=1)
df_resposta_obj = df_resposta_obj.drop(['Nota Final', 'Situação', 'Classificação', 'Convocado para correção\rda prova\rdiscrusiva/redação', 'Unnamed: 0', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8'], axis=1)


df_resposta_obj['Tipo_Prova'] = df_resposta_obj['Pontuação final nas Áreas de Conhecimento']
df_resposta_obj['Nota_obj'] = df_resposta_obj['Unnamed: 1']
df_resposta_obj['Situação_obj'] = df_resposta_obj['Unnamed: 2']
df_resposta_obj['Ranking_obj'] = df_resposta_obj['Unnamed: 3']

df_resposta_obj = df_resposta_obj.drop(['Pontuação final nas Áreas de Conhecimento', 'Tipo de\rProva', 'Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3'], axis=1)
df_resposta_obj = df_resposta_obj[1:] # Remove a primeira linha de cada tabela
df_resposta_obj['Incrição'] = df_resposta_obj['Incrição'].astype(str)
df_resposta_obj['Incrição'] = df_resposta_obj['Incrição'].str.replace('.0', '').astype(object)
df_resposta_obj['Incrição'] = df_resposta_obj['Incrição'].str.zfill(10)

# Display the new DataFrame
df_resposta_obj
```

###Coletando apenas candidatos de Analista de TI

```python
# Exibe as linhas onde a coluna 'Cargo' contém 'Analista de Tecnologia da Informação'
analistas_ti_obj = df_resposta_obj[df_resposta_obj['Ocupação'].str.contains('ANALISTA DE TECNOLOGIA DA INFORMACAO', na=False)]
analistas_ti_obj
```

#Mesclando notas de candidatos de TI para apresentar Nota Objetiva e Discursiva

```python
# Merge the DataFrames
analistas_ti_geral = pd.merge(analistas_ti_obj, aprovados_ti_disc, left_on='Incrição', right_on='Num de inscrição', how='left')
analistas_ti_geral
```

###Reduzindo dados para coletar apenas os necessários para análise

```python
resultado_geral_1 = analistas_ti_geral.drop(['Código\rdo Cargo', 'Ocupação', 'Tipo_Prova', 'Num de inscrição', 'Nome Candidato', 'Ocupação'], axis=1)
resultado_geral_1
```

```python
# Remove rows where 'NPD' is null
resultado_geral_1 = resultado_geral_1.dropna(subset=['NPD'])
resultado_geral_1
```

```python
resultado_geral_1.info()
```

```python
resultado_geral_join = pd.DataFrame(resultado_geral_1)

# Convert 'NPD' and 'Nota_obj' columns to numeric, handling errors
resultado_geral_join['NPD'] = resultado_geral_1['NPD'].str.replace(',', '.').astype(float)
#resultado_geral_join['NPD'] = pd.to_numeric(resultado_geral_1['NPD'])
resultado_geral_join['Nota_obj'] = pd.to_numeric(resultado_geral_1['Nota_obj'], errors='coerce')

# Calculate 'Nota_Geral'
resultado_geral_join['Nota_Geral'] = resultado_geral_join['Nota_obj'] + resultado_geral_join['NPD']
resultado_geral_join
```

##Classificação final

```python
# prompt: Gere um resultado como classificação decrescente com base na  coluna 'Nota_Geral', apresentando apenas as colunas de 'Nome do Candidato', 'Nota_obj', 'NPD' e 'Nota_Geral', e acrescente a coluna de index apresentando em qual posição do ranking o candidato está

# Sort the DataFrame by 'Nota_Geral' in descending order
resultado_geral_join_sorted = resultado_geral_join.sort_values(by='Nota_Geral', ascending=False)

# Reset the index to create a rank column
resultado_geral_join_sorted = resultado_geral_join_sorted.reset_index(drop=True)

# Rename the index to 'Ranking'
resultado_geral_join_sorted.index.name = 'Ranking'

# Select the desired columns and display the result
resultado_final = resultado_geral_join_sorted[['Nome do Candidato', 'Nota_obj', 'NPD', 'Nota_Geral']]
resultado_final
```
